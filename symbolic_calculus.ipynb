```python
_S = show
_b = bool

def _C(f):
    '''
       canonicalize_radical == simplify
    '''
    return f.canonicalize_radical()

def _cof(f, *args):
    return f.coefficient(*args)

def _Cg(ci):
    return ci.conjugate()

def _not(f, g):
    return f.contradicts(g)
```
